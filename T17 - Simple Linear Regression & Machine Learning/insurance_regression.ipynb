```python
#Importing the relevant modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import the insurance data

df = pd.read_csv("insurance.csv")
df.head()
```

```python
# Using iloc function to select rows from the dataframe for plotting
# age is an independent variable while charges is a dependent variable.
# I suspect that there is a relationship between the 2 variables and will perform the following analysis.

age = df.iloc[:,:1].values
charges = df.iloc[:,-1].values
```

```python
# Plotting the scatterplot with age on X-Axis and charges on Y-Axis
plt.scatter(age,charges,color = 'b')
plt.xlabel('Years')
plt.ylabel('Charges')
plt.show()
```

```python
# Creating an empty linear regression model
charges_model = LinearRegression()

# Creating the model using fit (age, charges)
charges_model.fit(age,charges)
```

```python
#Use model to make a prediction
charges_pred=charges_model.predict(age)

```

```python
# charges model coefficients
print('Coefficients: \n', charges_model.coef_)
```
