```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df=pd.read_csv('balance.txt',delim_whitespace=True)
```

Q1: Comparing the average income based on Ethnicity:

```python
df.columns
```

```python
#print(df.groupby(['Ethnicity']).mean(numeric_only=True))
print("The average income of an African American is: ",df[df.Ethnicity=="African American"].loc[:,"Income"].mean())
print("The average income of a Caucasian is: ",df[df.Ethnicity=="Caucasian"].loc[:,"Income"].mean())
print("The average income of an Asian is: ",df[df.Ethnicity=="Asian"].loc[:,"Income"].mean())
```

Q2: On average, do married or single people have a higher balance?

```python
print("The balance of married people is: ",df[df.Married=="Yes"].loc[:,"Balance"].mean())
print("The balance of single people is: ",df[df.Married=="No"].loc[:,"Balance"].mean())

if df[df.Married=="Yes"].loc[:,"Balance"].mean() > df[df.Married=="No"].loc[:,"Balance"].mean():
    print("\nThe results suggest married people have a higher balance")
else:
    print("\nThe results suggest single people have a higher balance")
```

Q3: What is the highest income in our dataset?

```python
print("The highest income in our dataset is: ",df["Income"].max())
```

Q4: What is the lowest income in our dataset?

```python
print("The lowest income in our dataset is: ",df["Income"].min())
```

Q5: How many cards do we have recorded in our dataset?

```python
print("Total number of cards recorded in our dataset: ",df["Cards"].sum())
```

Q6: How many females do we have information for vs how many males?

```python
print("Total number of females is: ",df[df.Gender=="Female"].loc[:,"Gender"].count())
print("Total number of males is: ",df[df.Gender=="Male"].loc[:,"Gender"].count())
```
